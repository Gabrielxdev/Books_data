```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset= "Book Name", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribuition of Publishing Year")
plt.show()
#Podemos ver que a maioria dos livros foram publicados entre 2000 e 2020, com um pico em 2010.
```

```python
df["genre"].value_counts().plot(kind='bar')
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
# Podemos ver que o gênero mais popular é Fiction, seguido por Non-Fiction e Mystery.
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending= False)
# Podemos ver que o autor mais bem avaliado é Bill Watterson, seguido por J. R.R. Tolkien
```

```python
sns.boxplot(x="genre", y = "Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book Ratings Count by Genre")
plt.show()
#Podemos ver que o gênero com mais avaliações é Ficção, seguido por não ficção e kids.
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units sold")
plt.show()
#Podemos ver que o preço de venda tem uma correlação negativa com o número de unidades vendidas, ou seja, quanto mais caro o livro, menos unidades são vendidas.
```

```python
language_counts = df["language_code"].value_counts()
```

```python


plt.figure(figsize=(7, 8))
plt.pie(language_counts, labels = language_counts.index, startangle=90, autopct = '%1.1f%%', labeldistance=1.1);
plt.show()
# Podemos ver que a maioria dos livros estão em inglês
```

```python
df.columns
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
# Podemos ver que a editora com maior receita é Penguin Group(USA) LLC
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False).head(1)
# Podemos ver que a classificação de autor com a maior média de quantidade de avaliações é o autor com classificação intermediate
```

```python
df.groupby("language_code").size().sort_values(ascending=False)
# Podemos ver que a maioria dos livros estão em inglês novamente
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].var()
```

```python
plt.scatter(df["Book_average_rating"], df['Book_ratings_count'])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Scatter Plot of Book Average Rating vs Book Ratings Count")
plt.show()
```

```python
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending=False).head(10).plot(kind = 'bar')
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales by Author")
plt.show()
#Podemos ver que o autor com mais vendas é Harper Lee
```

```python
sns.boxplot(x="Author_Rating", y = "units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of Units Sold for Each Author Rating")
plt.show()
# Podemos ver que o terceiro quartil Excellent está muito superior em vendas unitárias em relação aos outros, menos de 75% das classificação dos autores tiveram vendas unitárias entre 30 mil ou menos, porém a classificação de outhor Novice possui uma mediana melhor entre 4 mil há 0
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind = 'line', marker = 'o')
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold Over the Years")
plt.show()
# Podemos ver que as vendas unitárias aumentaram ao longo dos anos, com um pico em 2015, e depois uma queda.
```
